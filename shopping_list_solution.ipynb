```python
from IPython.display import clear_output

shopping_list = []
def add_item(item):
    item_upper = str(item).upper()
    shopping_list.append(item_upper)

def remove_item(item):
    item_upper = str(item).upper()
    while item_upper in shopping_list:
        shopping_list.remove(item_upper)

def view_list():
    if total_items():
        for it in shopping_list:
            print(it)
    else:
        print("Shopping List is empty")

def calculate_total_cost():
    total_cost = 0
    for item in shopping_list:
        unit_price = float(input(f"Enter unit price for {item}: "))
        quantity = shopping_list.count(item)
        item_cost = unit_price * quantity
        total_cost += item_cost
    return total_cost

def total_items():
    return len(shopping_list)

def user_interface():
    print("Shopping List App")
    print("Your options are:\n")
    print("_______________________")
    print("1. ADD ITEM")
    print("2. REMOVE ITEM")
    print("3. VIEW ITEMS")
    print("4. TOTAL COST")
    print("5. QUIT APP")
    print("_______________________\n")

    choice = int(input("ENTER: "))

    # This code will clear the output of the current code cell
    clear_output()

    match choice:
        case 1:
            print("Adding item...")
            item = input("ENTER ITEM:")
            add_item(item)
            print(f"Added {item} to the basket")
        case 2: 
            print("Removing item...")
            item = input("ENTER ITEM:")
            success = remove_item(item)
            print(f"Removed {item} from the basket")
        case 3:
            print("Viewing list of items...")
            print("_____SHOPPING LIST_____")
            print("_______________________")
            view_list()
            print("_______________________\n\n")
        case 4: 
            cost = calculate_total_cost()
            print(f"Total cost: {cost}")
        case 5: 
            print("Exiting program...")
            return 1 
        case _:
            print("INVALID OPTION!") 

    return 0 

while True:
    try:
        control_code = user_interface()
        if (control_code):
            break 
    except Exception as e:
        print(f"ERROR: {e}")
        break
```
